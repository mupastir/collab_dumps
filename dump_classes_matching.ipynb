```
import json
import copy
import pytz
import pandas as pd
```

```
classes_matching = pd.read_excel('/content/Program_Relevance_Scoring_-_23_May_2020.xlsx', 'Classes Matching')
```

```
classes_matching.head(5)
```

```
classes_matching_df = classes_matching.rename(columns={"Capacity Provider Class Entered Below\nCapacity Seeker Listed to Right": "name"})
```

```
classes_matching_df = classes_matching_df.set_index('name')
```

```
classes_matching_df_stacked = classes_matching_df.stack()
```

```
classes_matching_df_stacked.head()
```

```
classes_matching_dict_v1 = {}
for key, value in classes_matching_df_stacked.to_dict().items():
  if classes_matching_dict_v1.get(key[0]):
    classes_matching_dict_v1[key[0]].update({key[1]: value})
  else:
    classes_matching_dict_v1.update({key[0]: {key[1]: value}})
```

```
with open('classes-matching.json', 'w+') as openfile:
    openfile.write(json.dumps(classes_matching_dict_v1, indent=1, sort_keys=True))
```
