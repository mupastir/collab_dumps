```
import json
import copy
import pytz
import pandas as pd
```

```
teritories_matching = pd.read_excel('/content/Program_Relevance_Scoring_-_23_May_2020.xlsx', 'Teritories Matching')
```

```
teritories_matching.head(5)
```

```
teritories_matching_df = teritories_matching.rename(columns={"Capacity Provider Territory Entered Below\nCapacity Seeker Listed to Right": "name"})
```

```
teritories_matching_df = teritories_matching_df.set_index('name')
```

```
teritories_matching_df_srtacked = teritories_matching_df.stack()
```

```
teritories_matching_dict = teritories_matching_df_srtacked.to_dict()
```

```
teritories_matching_dict = {"|".join(key): value for key, value in teritories_matching_dict.items()}
```

```
fixture_structure = {
  "model": "offers.offer",
  "pk": 500000,
  "fields": {
    "created_at": "2020-04-27T16:12:41.132Z",
    "updated_at": "2020-04-27T16:12:41.659Z",
    "is_featured": False,
    "is_hidden": False,
    "meta_title": "",
    "meta_description": "",
    "meta_author": "",
    "meta_keywords": "",
    "meta_link_canonical": "",
    "meta_og_locale": "",
    "meta_og_type": "",
    "meta_og_title": "",
    "meta_og_description": "",
    "meta_og_site_name": "",
    "meta_og_website_tags": "",
    "meta_og_image": "",
    "meta_og_image_secure_url": "",
    "meta_og_image_width": "",
    "meta_og_image_height": "",
    "meta_twitter_card": "",
    "meta_twitter_description": "",
    "meta_twitter_title": "",
    "meta_twitter_image": "",
    "offer_definition_uid": 23,
    "slug": None,
    "data": {
        "match_name": "",
        "match_value": 0.0
    },
    "profile": None,
    "parent": None,
    "lft": 1,
    "rght": 2,
    "tree_id": 6,
    "level": 0
  }
}
```

```
fixture_list = []
for key, value in teritories_matching_dict.items(): 
    new_item = copy.deepcopy(fixture_structure)
    new_item["pk"] = new_item["pk"] + 1
    new_item["fields"]["data"]["match_name"] = key
    new_item["fields"]["data"]["match_value"] = value
    fixture_structure = copy.deepcopy(new_item)
    fixture_list.append(new_item)
    fixture_structure = copy.deepcopy(new_item)
    fixture_list.append(new_item)
    

with open('offers-territories-matching.json', 'w+') as openfile:
    openfile.write(json.dumps(fixture_list, indent=1, sort_keys=True))
```
